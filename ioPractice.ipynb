**묵시적 형변환**<br>
컴파일러에서 자료의 손실의 최소화하여 크기가 작은 자료형에서 큰 자료형으로 변환한다.

**명시적 형변환**  

크기가 큰 자료형에서 작은 자료형으로 변환하면 Bit 손실이 발생한다. 이러한 결과를 막기위해 의도적으로 형을 변환할 필요가 있다.

```python
# 묵시적 형변환
#num1 정수
num1 = 10
print("num1 : ",type(num1))
#num2 정수
num2 = 3
print("num2 : ",type(num2))
#result = num1 + num2
result = num1 + num2
print("result : ",type(result))
```

```python
# 묵시적 형변환
#num1 정수
num1 = 10
print("num1 : ",type(num1))
#num2 실수
num2 = 3.3
print("num2 : ",type(num2))
#result = num1 + num2 실수
result = num1 + num2
print("result : ",type(result))
```

```python
# 명시적 형변환
# num1 정수
num1 = 10
print("num1 : ",type(num1))
#num2 정수
num2 = 3.3
print("num2 : ",type(num2))
#result = num1 + num2
result = num1 + num2
#******#
result = int(result)
#******#
print("result : ",type(result))
```
